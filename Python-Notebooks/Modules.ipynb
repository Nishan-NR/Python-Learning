```python
#Enum module
class Color(Enum):
    RED = 1
    GREEN = 2
    BLUE = 3

print(Color.RED)
print(Color(2))
print(Color['RED'])
```

```python
#DATE time module

time = datetime.datetime.now()
print(time)
print(time.year)
print(time.strftime("%M"))
print(time.strftime("%A"))
print(time.strftime("%B"))
print(time.strftime("%Y-%m-%d %H:%M:%S"))
```
