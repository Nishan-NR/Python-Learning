```python
mylist = ['box','table','chair','laptop','keyboard','dustbin','mouse']
print(mylist)
print(len(mylist)) # prints the number of item inside the mylist

mixed_list = ['strings',3,True,] # A list can coontain multiple types of data 
print(mixed_list)
```

```python
# Accessing list items
print(mylist[1]) # this is positive indexing which starts form 0 index 
print(mylist[-1]) # This is a negative indexing which starts form the last item
print(mylist[2:4]) # starts from index 2 and ends in index 3 (range)
print(mylist[:5])
print(mylist[:-1]) # this line prients all items except for last item
print(mylist[-5:4])
```

```python
# modifying list items

mylist[0] = 'first'       # Adding new items to list  
print(mylist)
mylist[2:3]= 'rock','bugs'
print(mylist)
mylist.append('sky') # append add item to last index
mylist.insert(0,'apple') # .insert helps us to insert item to desired index place
print(mylist)

mylist.remove('rock')
del mylist[1:2]     #Using inex to remove list item
print(mylist)
mylist.pop(2)       # .pop removes index item
mylist.pop()        # if you don't specify index it removes last item
print(mylist)
# To remove all item from a list use list_name.clear() 
# to delete entire list del list_name
```

```python
# loop in list
for i in mylist:        # prints each item in the list line by line
    print(i)
```

```python
# List comprehension, meaning: creating a new list form a existing list 
newlist = []

for i in mylist:
    if "t" in i:
        newlist.append(i)
print(newlist)
# if you want to do this in one line then this is the way
newlist = [i for i in mylist if "t" in i]

# if you want to add item with first letter " " then this is the way
secondlist = [x for x in mylist if x.startswith("t")]
print(secondlist)

# if you want to add item using index then this is the wayyyy
indexlist= mylist[2:5] # By using string slicing 
print(indexlist)
#Butt if you want to do it using stirng slicing
new_list = [mylist[i] for i in range(2,5)]
```

```python
# sorting list 
mylist.sort() # this will sort in alphabetic order 
mylist.sort(reverse= True) # and this will reverse the order
```

```python
# copy list
demolist = mylist.copy()
demolist = list(indexlist)
demolist = newlist[:]
print(demolist)
```

```python
# joining list
list1 = [2,3,5,21]
list2 = [3,4,4]

list3 = list1 + list2
print(list3)
#another way
for i in list2:
    list1.append(i)
print(list1)

#another way
list1.extend(list2)
print(list1)
```
