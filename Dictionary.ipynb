```python
#Dictionary in python
mydictionary =  {"name ":"nishan","class":"human", "height":5.11} #Note: Space counts as character/white space
print(mydictionary["name "])
x = mydictionary.get("class") # the menthod get() will get the value of key
y = mydictionary.keys() # this gives us the list of keys
print(y)
print(x)
#updating the dictionary 
mydictionary["gender"] = "male"
print(mydictionary)

# if you want to assign multiple values to a key then this the way 
mydictionary["name "] = ("sunil","bimal")
print(mydictionary.get("name "))
```

```python
#you always can update your key there are differents way to achieve that some of them are 
mydictionary["height"] = 6
mydictionary.update({'study': "Bsc IT"}) #updating dictionary 
print(mydictionary)

#removing the item 
mydictionary.pop("class") # this will remove the key item 
mydictionary.popitem() # This will remove the last item in dictionary
print(mydictionary)
```

```python
demodict = mydictionary.copy() #this will create a copy, remember if you don't use copy() method then it will only change the name of dictionary 
#another way to copy dictionary is 
newdict = dict(mydictionary)
print(newdict)
print() # this will print empty line 

print(id(demodict))
print(id(mydictionary))
print()

demodict.clear() # clear method will clear all the item inside the dictionary
del demodict #   this will delete entire dictionary 
del mydictionary["height"]
print(mydictionary)
```
