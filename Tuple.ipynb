```python
mytuple = ("light","jacket","table","chair") # tuples are unchangable 
print(mytuple)


# accessing tuple items
print(mytuple[1])
print(mytuple[-1])
print(mytuple[2:4])

mytuple += "apple", 
print(mytuple)
```

```python
# in tuple there is no attribute named append but you can use it by this way
y = list(mytuple)
y.append("new_item")
mytuple = tuple(y)
print(mytuple)

x = mytuple.index("table") #.index() looks for a value not an index number
print(x)
```
